```julia
using CairoMakie, Random
include("makieTheme1.jl") # don't forget to include the theme. 

function t_qubit(, 1, 2, f, f)
    2 + 2* - cos(1) - cos(2) - 2**cos(*f)*cos(2**f + *f - 1- 2)
end
1 = 2 = LinRange(0, 4*, 100)
z = [t_qubit(.61, x, y, 0.2, .1) for x in 1, y in 2]
fig = Figure(resolution = (470, 400))
ax = Axis(fig, aspect = 1, xlabel = "1", ylabel = "2")
clines = contour!(1, 2, z, colormap = :plasma, levels = 20, linewidth = 1.5)
limits!(ax, 0, 4, 0, 4)
cbar  = Colorbar(fig, clines, label ="-q", labelpadding = 0)
fig[1, 1] = ax
fig[1, 2] = cbar
#save("./results/FigContour1.svg", scene, pt_per_unit = 0.7);
save("./results/FigContour1.png", fig, px_per_unit = 2)
fig
```
